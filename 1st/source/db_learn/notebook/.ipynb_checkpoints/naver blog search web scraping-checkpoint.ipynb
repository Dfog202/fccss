```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
naver_query = input("What do you wanna know??")
post_list_page = input("What page do you wanna go??")

url = "https://search.naver.com/search.naver?where=post&query=" + naver_query + "ie=utf8&start=" + post_list_page
```

```python
html = urllib.request.urlopen(url)
source = html.read()
html.close()
```

```python
soup = BeautifulSoup(source, 'html.parser')
print(soup)
```

```python
blog_list = soup.find(id="elThumbnailResultArea")
print(blog_list)
```

```python
all_dl = blog_list.find_all("dl")
print(all_dl)
```

```python
for dl in all_dl:
    post_title = dl.find("a", {'class':'sh_blog_title'})['title']
    blog_name = dl.find("a",{'class' :'txt84'}).text
    print(post_title," | ", blog_name)
```
