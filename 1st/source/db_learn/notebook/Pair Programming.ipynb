## 1. Stack OverFlow web scraping

```python
import urllib.request
from bs4 import BeautifulSoup as beausoup


url = "http://stackoverflow.com/questions/tagged/python"

html = urllib.request.urlopen(url)
source = html.read()
html.close()

soup = beausoup(source, 'html.parser')

q_list = soup.find(id="questions")

summaries = q_list.find_all("div", {"class":"summary"})

for summary in summaries:
    title = summary.find("h3").text
    user_id = summary.find("div", {"class":"user-details"}).a.text
    print(title," | " ,user_id)
```

## 2. Fizzbuzz

```python
def numguesser():
    user_num = int(input("What number do you like? "))
    
    if user_num > 0:
        for i in range(1,user_num + 1):
            if i % 3 == 0 and i % 5 == 0:
                print("Fizzbuzz")
            elif i % 3 == 0:
                print("Fizz")
            elif i % 5 == 0:
                print("Buzz")
            else:
                print(i)
    else:
        print("try again\n")
        return numguesser()

numguesser()
```

## 3. Numguess

```python
import random


answer = random.randint(1,100)
#Show answer for debug
print(answer)
name = input("Hi there, What is your name?? ")
print("Hi, ", name, "plz guess the number!!")

count = 10

# Type check
while True or count > 0:
    maxValue = 100
    try:
        print(count, "chances left.")
        guess = int(input())
    except ValueError:
        print("What R U doing?? guess the NUMBER!!!")
        count -= 1
        continue
    else:
        if guess < 1 or guess > maxValue:
            print("The number you guessed is too low or high!!!")
            count -= 1
            continue
        else:    
            if guess == answer:
                print("Correct!! ", 10 - count, "times you did!")
                break
            elif guess > answer:
                print("Too high!!")
                count -= 1
            else:
                print("Too Low!!")
                count -= 1
```
